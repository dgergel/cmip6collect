###  Create a response table using qgrid to pick the data and pivot table to format

```python
import numpy as np
import pandas as pd
import os
import gcsfs
import xarray as xr
```

```python
import pandas as pd
import qgrid
dfcat = pd.read_csv('https://cmip6.storage.googleapis.com/cmip6-zarr-consolidated-stores.csv')
```

```python
widget=qgrid.show_grid(dfcat)
widget #*historical, *ssp126, *ssp245, *ssp370, *ssp585
```

```python
# save results
df = widget.get_changed_df()     # use this to save filtered dataframe
# df = widget.get_selected_df()  # use this instead to save selected rows
df.experiment_id.unique(), len(df)
```

```python
df.variable_id.unique()
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

dm = df[['experiment_id','source_id','variable_id','member_id']].groupby([
         'experiment_id','source_id','variable_id']).nunique()[['member_id']]

table = pd.DataFrame.pivot_table(dm,
                                 values='member_id',
                                 index=['source_id','variable_id'],
                                 columns=['experiment_id'],
                                 aggfunc=np.sum,
                                 fill_value=0)
print('activity_id, table_id = ',df.activity_id.unique(),df.table_id.unique())
display(table)
```

```python
dm = df[['table_id','experiment_id','source_id','variable_id','member_id','dcpp_init_year']].groupby([
         'table_id','experiment_id','source_id','variable_id','dcpp_init_year']).nunique()[['member_id']]

table = pd.DataFrame.pivot_table(dm,
                                 values='member_id',
                                 index=['table_id','experiment_id','source_id','dcpp_init_year'],
                                 columns=['variable_id'],
                                 aggfunc=np.sum,
                                 fill_value=0)

print('current CMIP6 datasets')
display(table)
```
